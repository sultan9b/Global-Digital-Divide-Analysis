```python
import pandas as pd
```

#### 1. Internet Users processing

```python
# загрузка данных
internet_users = pd.read_excel('../data/02_manual_processed/internet_users.xlsx', sheet_name='Data')
internet_users
```

```python
# unpivot: превращение года в строки в столбце Year
internet_users_melted = pd.melt(
    internet_users,
    id_vars=['Country Name', 'Country Code'],
    var_name='Year',
    value_name = 'Internet_Users_Percent'
)

# конвертация в число
internet_users_melted.Year = pd.to_numeric(internet_users_melted.Year, errors='coerce')
```

```python
internet_users_melted.head(10)
```

```python
# удаление названия стран
internet_users_melted = internet_users_melted.drop(['Country Name'], axis=1)
```

```python
# сортировка по странам и годам
internet_users_melted = internet_users_melted.sort_values(['Year'])
internet_users_melted.head(10)
```

```python
# сохранение итогов
internet_users_melted.to_excel('../data/03_pandas_processed/internet_users_transformed.xlsx', sheet_name='internet_users', index=False)
```

```python
# строки со странами в таблице internet_users, с которыми сравнимаем строки в других таблицах и фильтруем по ним
countries_to_keep = internet_users_melted['Country Code'].unique()
print(f'Количество стран в internet_users: {len(countries_to_keep)}')
```

#### 2. GDP per capita processing

```python
# загрузка данных
gdp_per_capita = pd.read_excel('../data/01_raw/gdp per capita.xlsx', sheet_name='Data')
gdp_per_capita.head()
```

```python
# удаление не нужных столбцов
gdp_per_capita = gdp_per_capita.drop(['Country Name', 'Indicator Name', 'Indicator Code'], axis=1)
gdp_per_capita.head()
```

```python
# фильтруем и оставляем только те строки, что встречаются в таблице internet_users
gdp_per_capita_filtered = gdp_per_capita[
    gdp_per_capita['Country Code'].isin(countries_to_keep)
]

print(f'Кол-во строк до фильтрации: {len(gdp_per_capita)}')
print(f'Кол-во строк после фильтрации: {len(gdp_per_capita_filtered)}')
```

```python
# unpivot: превращение года в строки в столбце Year
gdp_per_capita_melted = pd.melt(
    gdp_per_capita_filtered,
    id_vars=['Country Code'],
    var_name='Year',
    value_name = 'Gdp_Per_Capita'
)

# конвертация в число
gdp_per_capita_melted.Year = pd.to_numeric(gdp_per_capita_melted.Year, errors='coerce') 
```

```python
# сортировка по странам и годам
gdp_per_capita_melted = gdp_per_capita_melted.sort_values(['Country Code', 'Year'])
gdp_per_capita_melted.head(10)
```

```python
# сохранение итогов
gdp_per_capita_melted.to_excel('../data/03_pandas_processed/gdp_per_capita_transformed.xlsx', sheet_name='gdp per capita', index=False)
```

### Таблица Regions

```python
# загрузка данных
regions = pd.read_excel('../data/02_manual_processed/internet_users.xlsx', sheet_name='Metadata - Countries')

# для определения континентов используется другая таблица
# данные взяты по адресу https://github.com/lukes/ISO-3166-Countries-with-Regional-Codes/blob/master/all/all.csv
all = pd.read_csv('../data/01_raw/all.csv')
```

```python
regions
```

```python
# удаление не нужных стоблцов
regions = regions.drop(['SpecialNotes'], axis=1)
regions.head()
```

```python
all.head()
```

```python
# удаление ненужных столбцов
all = all.drop(
    ['name',
    'alpha-2', 
    'country-code', 
    'iso_3166-2', 
    'intermediate-region', 
    'region-code', 
    'sub-region-code', 
    'intermediate-region-code'],
    axis=1
    )
all.head()
```

```python
# проверка на пустые значения
regions.isnull().sum()
```

```python
# удаление пустых значении в столбца Region
regions.dropna(subset=['Region'], inplace=True)
print(regions.shape, regions.isnull().sum(), sep='\n\n')
```

```python
# проверка размеров таблиц для соединения
print(regions.shape, all.shape, sep=' ')
```

```python
# объединение двух таблиц
regions_merged = pd.merge(
    regions,
    all,
    left_on='Country Code',
    right_on='alpha-3',
    how='left'
)
regions_merged
```

```python
# удаление ненужных столбцов и переименование остальных
regions_merged = regions_merged.drop([
    'Region',
    'alpha-3'
    ], axis=1)
regions_merged.rename(columns={'region':'Continent', 'TableName':'Country Name', 'sub-region':'Region'}, inplace=True)
regions_merged.sort_values('Country Code')
regions_merged.head(10)
```

```python
# фильтруем и оставляем только те строки, что встречаются в таблице internet_users
regions_filtered = regions_merged[
    regions_merged['Country Code'].isin(countries_to_keep)
]

print(f'Кол-во строк до фильтрации: {len(regions_merged)}')
print(f'Кол-во строк после фильтрации: {len(regions_filtered)}')
```

```python
# сохранение итогов
regions_filtered.to_excel('../data/03_pandas_processed/regions.xlsx', sheet_name='regions', index=False)
```

### Объединенная таблица

```python
# проверка размеров всех таблиц
print(internet_users_melted.shape,
    gdp_per_capita_melted.shape, 
    population_data_melted.shape, 
    mobile_cellular_melted.shape, 
    regions_filtered.shape)
```

```python
# объединение internet_users и population_data
merged_data = pd.merge(
    internet_users_melted,
    population_data_melted,
    on=['Country Code', 'Year'],
    how='left',
    suffixes=('', '_pop')
)
```

```python
# добавляем gdp_per_capita
merged_data = pd.merge(
    merged_data,
    gdp_per_capita_melted,
    on=['Country Code', 'Year'],
    how='left',
    suffixes=('', '_gdp')
)
```

```python
# добавляем mobile_cellular_subs
merged_data = pd.merge(
    merged_data,
    mobile_cellular_melted,
    on=['Country Code', 'Year'],
    how='left',
    suffixes=('', '_mobile')
)
```

```python
merged_data.sort_values(['Country Code', 'Year'])
```

```python
# сохранение итогов
merged_data.to_excel('../data/03_pandas_processed/merged_data.xlsx', sheet_name='data', index=False)
```
